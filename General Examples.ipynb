```ruby
require 'daru/view'
```

```ruby
Daru::View.plotting_library = :highcharts
```

```ruby
opts = {
      chart: {
        map: 'custom/europe',
        borderWidth: 1
      },
      title: {
          text: 'Nordic countries'
      },

      subtitle: {
          text: 'Demo of drawing all areas in the map, only highlighting partial data'
      },

      legend: {
          enabled: false
      }
    }

series_dt = [
  {
    name: 'Country',
    data: [
        ['is', 1],
        ['no', 1],
        ['se', 1],
        ['dk', 1],
        ['fi', 1]
    ],
    dataLabels: {
        enabled: true,
        color: '#FFFFFF',
        formatter: 'function () {
            if (this.point.value) {
                return this.point.name;
            }
        }'.js_code
    }
  }
]

highlight = Daru::View::Plot.new(series_dt, opts, chart_class: 'map')
highlight.show_in_iruby
```
